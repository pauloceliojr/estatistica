```python
ax = sns.histplot(notas1.rating, kde=True, stat="density")
ax.figure.set_size_inches(14, 6)
ax.set_title("Distribuição das notas para o Toy Story", fontsize=18)
ax.set_xlabel("Nota", fontsize=12)
ax.set_ylabel("Densidade", fontsize=12)
```

```python
ax = sns.boxplot(x="rating", data=notas1, orient="h")
ax.figure.set_size_inches(14, 6)
ax.set_title("Distribuição das notas para o Toy Story", fontsize="18")
ax.set_xlabel("Nota", fontsize=12)
```

```python
notas1.rating.mean()
```

```python
notas1.rating.count()
```

```python
zconfint(notas1.rating) # mais de 30 amostras (215)
```

```python
from statsmodels.stats.weightstats import ztest
```

```python
stat, p_valor = ztest(notas1.rating, value=3.4320503405352603)
print(stat)
print(p_valor.round(2))
```

Conclusão: o filme Toy Story tem uma média diferente de todos os filmes.

```python
temp = notas1.rating.sample(frac=1, random_state=75241) #pega, aleatoriamente (random_state), todos os registros do conjunto de dados (frac)

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p_valor = ztest(temp[0:i], value=3.4320503405352603)
    return (i, media, p_valor) # número de amostras, média, p-valor

valores = np.array([calcula_teste(i) for i in range(2, len(temp))]) # sobre o range iniciando em 2: primeira linha só trouxe um elemento e trouxe o p-valor como nulo porque a amostra só tinha um registro

ax = sns.lineplot(x=valores[:,0], y=valores[:,1])
ax = sns.lineplot(x=valores[:,0], y=valores[:,2])
ax.hlines(y=0.05, xmin=2, xmax=len(temp), colors="r")
ax.figure.set_size_inches(14, 6)
```

# Comparação de dois conjuntos de amostras

```python
print(ztest(notas1.rating, notas.rating))
zconfint(notas1.rating, notas.rating)
```

```python
print(ztest(notas.rating, notas1.rating))
zconfint(notas.rating, notas1.rating)
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(notas.rating, notas1.rating)
```

```python
descr_todas_as_notas = DescrStatsW(notas.rating)
descr_toystory = DescrStatsW(notas1.rating)
comparacao = descr_todas_as_notas.get_compare(descr_toystory)
comparacao.summary(use_t=False)
```

```python
comparacao.summary() # t-test é padrão
```
